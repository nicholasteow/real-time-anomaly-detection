```python
import joblib
import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
best_model = joblib.load('Created_files/best_isolation_forest_model.pkl')
best_model
```

```python
no_psu_with_fake_data_df_test = pd.read_pickle('Created_files/no_psu_with_fake_data_df_test.pkl')
no_psu_with_fake_data_df_test
```

```python
# Get 100 random rows where is_anomaly_truth is 1
non_anomalies = no_psu_with_fake_data_df_test[no_psu_with_fake_data_df_test['is_anomaly_truth'] == 1].sample(n=100, random_state=42)
non_anomalies
```

```python
# Get 100 random rows where is_anomaly_truth is -1
anomalies = no_psu_with_fake_data_df_test[no_psu_with_fake_data_df_test['is_anomaly_truth'] == -1].sample(n=100, random_state=42)

anomalies
```

```python
# Define the features used for prediction
features = ['I_MEAS', 'TC_LD', 'TC_CMB', 'TC_CPS', 'PD2']

# Extract the feature data
new_X = non_anomalies[features]

# Preprocess the new data
new_X_preprocessed = best_model.named_steps['preprocessor'].transform(new_X)

# Predict anomaly labels with the best model
new_anomaly_labels = best_model.named_steps['model'].predict(new_X_preprocessed)

# Add the predictions to the new data
non_anomalies['is_anomaly_pred'] = new_anomaly_labels

non_anomalies['is_anomaly_pred'].value_counts()
non_anomalies
```

```python
# Define the features used for prediction
features = ['I_MEAS', 'TC_LD', 'TC_CMB', 'TC_CPS', 'PD2']

# Extract the feature data
new_X = anomalies[features]

# Preprocess the new data
new_X_preprocessed = best_model.named_steps['preprocessor'].transform(new_X)

# Predict anomaly labels with the best model
new_anomaly_labels = best_model.named_steps['model'].predict(new_X_preprocessed)

# Add the predictions to the new data
anomalies['is_anomaly_pred'] = new_anomaly_labels

anomalies['is_anomaly_pred'].value_counts()
anomalies
```

```python
truth_labels = non_anomalies['is_anomaly_truth']
predicted_labels = non_anomalies['is_anomaly_pred']

# Creating the confusion matrix with labels [1, -1] where 1 = no anomaly, -1 = anomaly
conf_matrix = confusion_matrix(truth_labels, predicted_labels, labels=[1, -1])

print("Confusion Matrix:")
print(conf_matrix)

# True Positives (TP) --> No issues, predict no issues
# False Positives (FP) --> Anomalies, predict no issues
# False Negatives (FN) --> No issues, predict anomalies
# True Negatives (TN) --> Anomalies, predict anomalies
TP = conf_matrix[0, 0]
FP = conf_matrix[1, 0]
FN = conf_matrix[0, 1]
TN = conf_matrix[1, 1]

FPR = FP / (FP + TN) if (FP + TN) != 0 else 0
FNR = FN / (FN + TP) if (FN + TP) != 0 else 0
accuracy = accuracy_score(truth_labels, predicted_labels)


classification_rep = classification_report(truth_labels, predicted_labels, target_names=['anomaly', 'no anomaly'], zero_division=0)
print("\nFalse Positive Rate (FPR):", FPR)
print("False Negative Rate (FNR):", FNR)
print("Accuracy:", accuracy)
print("\nClassification Report:")
print(classification_rep)
```
